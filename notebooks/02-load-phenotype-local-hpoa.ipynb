# 02 - Load `phenotype_local.hpoa` into Pandas

## Learning objectives
- Locate the `phenotype_local.hpoa` file in the project.
- Read a TSV file into a pandas DataFrame.
- Perform quick sanity checks on the loaded data.

```python
from pathlib import Path
import pandas as pd
```

```python
# Resolve project root so the notebook works from both project root and notebooks/.
cwd = Path.cwd()
if (cwd / "data" / "phenotype_local.hpoa").exists():
    project_root = cwd
elif (cwd.parent / "data" / "phenotype_local.hpoa").exists():
    project_root = cwd.parent
else:
    raise FileNotFoundError("Could not find data/phenotype_local.hpoa from current working directory.")

hpoa_path = project_root / "data" / "phenotype_local.hpoa"
print(f"Reading file: {hpoa_path}")
```

```python
# The .hpoa file is tab-separated, so we use sep='\t'.
# dtype='string' keeps IDs/text values consistent and avoids mixed-type warnings.
df_hpoa = pd.read_csv(hpoa_path, sep='\t', dtype='string')

print(f"Rows: {df_hpoa.shape[0]:,}")
print(f"Columns: {df_hpoa.shape[1]}")
df_hpoa.head()
```

```python
# Optional quick checks for class discussion.
print("Column names:")
print(df_hpoa.columns.tolist())

print("\nMissing values per column (top 10):")
display(df_hpoa.isna().sum().sort_values(ascending=False).head(10))

# Check if some columns are completely empty (all values are missing).
empty_columns = df_hpoa.columns[df_hpoa.isna().all()].tolist()
if empty_columns:
    print("\nColumns with no values:", empty_columns)
else:
    print("\nNo columns are completely empty.")
```
